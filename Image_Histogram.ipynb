Import libary yang dibutuhkan.

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

Muat gambar asli.

```python
image_path = 'compressed profile.png'

img = cv.imread(image_path)

img = cv.resize(img,(720, 900))

cv2_imshow(img)
```

Ubah ke grayscale.

```python
img = cv.cvtColor(img,cv.COLOR_BGR2GRAY)

cv2_imshow(img)
```

Plot histogram.

```python
plt.hist(img.ravel(),256,[0,256]); plt.show()
```

Dari hasil plot histogram tersebut, dapat disimpulkan bahwa foto saya termasuk ke dalam citra gelap (**under exposed**).
